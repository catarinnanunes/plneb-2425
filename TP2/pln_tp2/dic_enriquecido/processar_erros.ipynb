```python
import json

with open('dicionario_atualizado1.json', 'r', encoding='utf-8') as f:
    dicionario = json.load(f)

chaves_normalizadas = {}
duplicados = []

for chave in dicionario:
    chave_lower = chave.lower()
    print(chave)  
    if chave_lower in chaves_normalizadas:
        duplicados.append((chave, chaves_normalizadas[chave_lower]))
    else:
        chaves_normalizadas[chave_lower] = chave

print(f"\nNúmero de chaves únicas (ignorando maiúsculas/minúsculas): {len(chaves_normalizadas)}")
print(f"Número de duplicados encontrados: {len(duplicados)}")

for dup in duplicados:
    print(f"Duplicado: '{dup[0]}' e '{dup[1]}'")
```

```python
import json

# Função que capitaliza só a primeira letra da chave
import json

def capitalizar_chaves(d):
    novo_dict = {}
    for chave, valor in d.items():
        nova_chave = chave[0].upper() + chave[1:]

        # Se já existe uma entrada com a chave capitalizada, junta os conteúdos
        if nova_chave in novo_dict:
            if isinstance(valor, dict) and isinstance(novo_dict[nova_chave], dict):
                novo_dict[nova_chave] = {**novo_dict[nova_chave], **valor}
            else:
                # Em caso de lista de entradas, junta as listas
                if isinstance(novo_dict[nova_chave], list) and isinstance(valor, list):
                    novo_dict[nova_chave].extend(valor)
                else:
                    # Se forem tipos diferentes ou conflitantes, cria uma lista com ambos
                    novo_dict[nova_chave] = [novo_dict[nova_chave], valor]
        else:
            novo_dict[nova_chave] = valor
    return novo_dict

with open('dicionario_atualizado1.json', 'r', encoding='utf-8') as f:
    dicionario = json.load(f)

dicionario_fundido = capitalizar_chaves(dicionario)

with open('dicionario_juntar.json', 'w', encoding='utf-8') as f:
    json.dump(dicionario_fundido, f, ensure_ascii=False, indent=4)
```
